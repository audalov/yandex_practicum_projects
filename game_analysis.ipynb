````python
def print_missing_values_count(column_name, column_label):
    isnull_count = data[column_name].isnull().sum()
    total_count = len(data)
    percentage = isnull_count / total_count
    output = f"–ö–æ–ª–∏—á–µ—Å—Ç–≤–æ —è—á–µ–µ–∫ —Å –ø—Ä–æ–ø—É—â–µ–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –≤ —Å—Ç–æ–ª–±—Ü–µ '{column_label}': {isnull_count} (–¥–æ–ª—è –æ—Ç –æ–±—â–µ–≥–æ –º–∞—Å—Å–∏–≤–∞: {percentage:.2%})"
    print(output)

print_missing_values_count('user_score', 'user_score')
print_missing_values_count('critic_score', 'critic_score')
````

–ü—Ä–æ–ø—É—â–µ–Ω–Ω—ã—Ö –∑–Ω–∞—á–µ–Ω–∏–π –≤ —Å—Ç–æ–ª–±—Ü–∞—Ö user_score –∏ critic_score –º–Ω–æ–≥–æ.

````python
missing_percentage_user_score = data.groupby('year_of_release')['user_score'].apply(lambda x: x.isnull().mean())
missing_percentage_critic_score = data.groupby('year_of_release')['critic_score'].apply(lambda x: x.isnull().mean())

total_records = data['year_of_release'].value_counts()

plt.style.use('seaborn-darkgrid')

fig, ax = plt.subplots(figsize=(18, 6))

ax.bar(missing_percentage_user_score.index, missing_percentage_user_score, width=0.4, label='–î–æ–ª—è –ø—Ä–æ–ø—É—Å–∫–æ–≤ user_score', color='tan', align='center')
ax.bar(missing_percentage_critic_score.index + 0.4, missing_percentage_critic_score, width=0.4, label='–î–æ–ª—è –ø—Ä–æ–ø—É—Å–∫–æ–≤ critic_score', color='purple', align='center')
ax.set_xlabel('–ì–æ–¥ –≤—ã–ø—É—Å–∫–∞', fontsize=12)
ax.set_ylabel('–î–æ–ª—è –ø—Ä–æ–ø—É—Å–∫–æ–≤', fontsize=12)
ax.set_title('–î–æ–ª—è –ø—Ä–æ–ø—É—Å–∫–æ–≤ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—å—Å–∫–∏—Ö –æ—Ü–µ–Ω–æ–∫ –∏ –æ—Ü–µ–Ω–æ–∫ –∫—Ä–∏—Ç–∏–∫–æ–≤', fontsize=16, fontweight='bold')
ax.legend(fontsize=10)
ax.grid(axis='y', linestyle='--', linewidth=0.5)

ax.set_xticks(missing_percentage_user_score.index)

plt.xticks(rotation=45)

plt.show()
````

````python
platforms = ['PS4', 'XOne', '3DS', 'WiiU', 'PC']
results = []

for platform in platforms:
    data_platform = data_actual.query('platform == @platform and user_score > 0')
    correlation_critic = data_platform['critic_score'].corr(data_platform['sum_sales'])
    correlation_user = data_platform['user_score'].corr(data_platform['sum_sales'])
    results.append([platform, correlation_user, correlation_critic])

data_results = pd.DataFrame(results, columns=['platform', 'corrlation_on_user_score', 'dependence_on_critic_score'])
print(data_results)
````

## –°–æ—Å—Ç–∞–≤–ª–µ–Ω–∏–µ –ø–æ—Ä—Ç—Ä–µ—Ç–∞ –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç —Ä–µ–≥–∏–æ–Ω–∞

### –û–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ —Ç–æ–ø-5 –ø–ª–∞—Ç—Ñ–æ—Ä–º –ø–æ –ø—Ä–æ–¥–∞–∂–∞–º –¥–ª—è –∫–∞–∂–¥–æ–≥–æ —Ä–µ–≥–∏–æ–Ω–∞

````python
data_actual_region = data_actual.groupby('platform')[['na_sales', 'eu_sales', 'jp_sales']].sum()

data_actual_region['na_share'] = data_actual_region['na_sales'] / data_actual_region['na_sales'].sum()*100
data_actual_region['eu_share'] = data_actual_region['eu_sales'] / data_actual_region['eu_sales'].sum()*100
data_actual_region['jp_share'] = data_actual_region['jp_sales'] / data_actual_region['jp_sales'].sum()*100

data_actual_region = data_actual_region.round(2)

data_actual_region_styled = data_actual_region.style.format({
    'na_share': '{:.2f}%',
    'eu_share': '{:.2f}%',
    'jp_share': '{:.2f}%'
})

data_actual_region_styled
````

````python
top5_platforms_na = data_actual_region['na_sales'].nlargest(5)
top5_platforms_eu = data_actual_region['eu_sales'].nlargest(5)
top5_platforms_jp = data_actual_region['jp_sales'].nlargest(5)

other_na_sales = data_actual_region['na_sales'].sum() - top5_platforms_na.sum()
other_eu_sales = data_actual_region['eu_sales'].sum() - top5_platforms_eu.sum()
other_jp_sales = data_actual_region['jp_sales'].sum() - top5_platforms_jp.sum()

platforms_na = top5_platforms_na.index
platforms_eu = top5_platforms_eu.index
platforms_jp = top5_platforms_jp.index

fig, axs = plt.subplots(1, 3, figsize=(20, 6))

region_titles = ['–î–æ–ª–∏ –ø—Ä–æ–¥–∞–∂ –≤ –°–µ–≤–µ—Ä–Ω–æ–π –ê–º–µ—Ä–∏–∫–µ', '–î–æ–ª–∏ –ø—Ä–æ–¥–∞–∂ –≤ –ï–≤—Ä–æ–ø–µ', '–î–æ–ª–∏ –ø—Ä–æ–¥–∞–∂ –≤ –Ø–ø–æ–Ω–∏–∏']

for i, (top5, other_sales, platforms, region_title) in enumerate(zip([top5_platforms_na, top5_platforms_eu, top5_platforms_jp], [other_na_sales, other_eu_sales, other_jp_sales], [platforms_na, platforms_eu, platforms_jp], region_titles)):
    labels = platforms.tolist() + ['Other']
    sizes = top5.tolist() + [other_sales]
    colors = plt.cm.tab20c.colors[:len(labels)]  # –ò—Å–ø–æ–ª—å–∑—É–µ–º —Ü–≤–µ—Ç–∞ –∏–∑ tab20c colormap
    wedges, texts, autotexts = axs[i].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors, wedgeprops=dict(width=0.25))
    axs[i].set_title(region_title, fontsize=16)
    for autotext in autotexts:
        autotext.set_fontsize(14)


plt.suptitle('–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ –ø—Ä–æ–¥–∞–∂ –∏–≥—Ä –¥–ª—è —Ç—Ä–µ—Ö —Ä–µ–≥–∏–æ–Ω–æ–≤ –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –ø–ª–∞—Ç—Ñ–æ—Ä–º—ã', y=1.05, fontsize=16, fontweight='bold')
plt.show()
````
